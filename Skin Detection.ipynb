```python
import cv2
import numpy as np
import math
import datetime
import time
import matplotlib.pyplot as plt
```

```python
vid = cv2.VideoCapture(0)
while True:
    b,frame = vid.read()
    frame = cv2.flip(frame,1)
    
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    l_r = np.array([0,38,80],dtype="uint8")
    u_r = np.array([50,255,255],dtype="uint8")
    skin = cv2.inRange(hsv,l_r,u_r)
    kernel = (11,11)
    skin = cv2.erode(skin,(1,1),iterations=1)
    skin = cv2.dilate(skin,kernel,iterations=2)
    
    skin = cv2.medianBlur(skin,15)
    
    
    if cv2.waitKey(1)& 0xFF==ord('q'):
        plt.imshow(skin)
        cv2.destroyAllWindows()
        break
  
    cv2.imshow("Video",skin)
```
